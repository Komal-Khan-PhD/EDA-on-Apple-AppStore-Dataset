#### 5.3.3 Defining each column


A comprehensive set of information about each app is provided in the dataset related to mobile applications (apps) available on  the Apple App Store. Here's a breakdown of each column:

1. `App_Id`: A unique identifier assigned to each mobile application in the dataset.
2. `App_Name`: The name/title of the mobile application.
3. `AppStore_Url`: The URL/link to the mobile application's page on the respective app store (e.g., Apple App Store or Google Play Store).
4. `Primary_Genre`: The primary category or genre of the app (e.g., Games, Productivity, Social Networking, etc.).
5. `Content_Rating`: The age or content rating assigned to the app indicating the appropriate audience (e.g., Everyone, Teen, Mature, etc.).
6. `Size_Bytes`: The size of the app in bytes, indicating the amount of storage space it occupies on a device.
7. `Required_IOS_Version`: The minimum iOS version required for the app to run on Apple devices.
8. `Released`: The date when the app was initially released or made available for download.
9. `Updated`: The date of the most recent update or version release of the app.
10. `Version`: The version number of the app (e.g., 1.0, 2.5.3, etc.).
11. `Price`: The price of the app, if not free.
12. `Currency`: The currency in which the app's price is listed.
13. `Free`: A flag or indicator whether the app is free or not (e.g., True/False).
14. `DeveloperId`: A unique identifier for the developer of the app.
15. `Developer`: The name of the developer or development company behind the app.
16. `Developer_Url`: The URL/link to the developer's profile or page.
17. `Developer_Website`: The URL/link to the developer's official website.
18. `Average_User_Rating`: The average rating given by users who have downloaded and used the app.
19. `Reviews`: The number of user reviews/ratings submitted for the app.
20. `Current_Version_Score`: The current average rating score for the most recent version of the app.
21. `Current_Version_Reviews`: The number of user reviews/ratings specific to the most recent version of the app.

Both Price and Size_Bytes are numeric column. so we can impute their missing values with mean.

```python
df['Price'] = df.Price.fillna(value=df['Price'].mean())
df['Size_Bytes'] = df.Size_Bytes.fillna(value=df['Size_Bytes'].mean())
```

```python
df.isnull().sum().sort_values(ascending = False)
```

2. Trend of P.G with time? which PG is progressing with time?

```python
df['Released'] = pd.to_datetime(df['Released']) 
# convert to datetime format
df['Year_Release'] = df['Released'].dt.strftime('%Y')
```

```python
# Assuming your DataFrame is named 'df' and the columns are named 'column1' and 'column2'
df_counts = df.groupby('Year_Release')["Primary_Genre"].value_counts().reset_index(name="count")
df_counts_5y = df_counts[df_counts["Year_Release"] >= "2015"]
# df_counts_5y = df_counts[df_counts["Year_Release"] >= "2015" & df_counts["count"] >= "2015"]
```

```python
df_counts_5y
```

Since we have very few data in year 2022 we'll only focus on the previous years.

```python
df_counts_5y = df_counts[df_counts['Year_Release'].between("2011", "2021")]
```

```python
# Assuming your DataFrame is named 'df' and the columns are named 'year', 'student', and 'score'
top_PG = df_counts_5y.groupby('Year_Release').apply(lambda x: x.nlargest(5, 'count')).reset_index(drop=True)

# Print the top students
print(top_PG)
```

```python
# # Plot with Plotly
fig = px.bar(top_PG, x='Year_Release', y='count', color='Primary_Genre', barmode='group')
fig.update_layout(title='Top 5 Primary_Genre for last few Years', xaxis_title='Year_Release', yaxis_title='count')
fig.show()
```
